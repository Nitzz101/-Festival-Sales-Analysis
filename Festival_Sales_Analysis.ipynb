<font size="8">Festival Sales Analysis</font>

```python
import pandas as pd 
import numpy as np
import seaborn as sb
```

```python
df = pd.read_csv('Festival_Sales_Analysis.csv', encoding='unicode_escape')
```

<font size="8">Explore the Data</font>

```python
df.head()
```

```python
df.columns
```

<font size="8">Data Cleaning</font>

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.rename(columns = {'Age Group' : 'Age_Group'}, inplace=True)
```

```python
df.columns
```

<font size="8">Data Visualization</font>

<font size="6">Gender</font>

```python
data = sb.countplot(x = 'Gender', data = df, )

for bars in data.containers:
    data.bar_label(bars)
```

```python
total_sales = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

values = sb.barplot(x='Gender', y='Amount', data=total_sales, hue='Gender')

#for bars in values.containers:
    #values.bar_label(bars)
```

<font size="6">Age Group</font>

```python
age_group = sb.countplot(x='Age_Group', data = df)

for bars in age_group.containers:
    age_group.bar_label(bars)
```

```python
sales_based_on_age_group = df.groupby(['Age_Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

total_amount = sb.barplot(x='Age_Group', y='Amount', data = sales_based_on_age_group)

#for bars in total_amount.containers:
    #total_amount.bar_label(bars)
```

<font size='6'>States</font>

```python
sales_based_on_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)

sb.set(rc={'figure.figsize':(15,5)})

sb.barplot(x='State', y='Orders', data = sales_based_on_states)
```

```python
sales_based_on_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)

sb.set(rc={'figure.figsize':(15,5)})

sb.barplot(x='State', y='Amount', data = sales_based_on_states)
```

<font size='6'>Marital Status</font>

```python
df.columns
#sb.countplot(x='')
```

```python
marital_status = sb.countplot(x='Marital_Status', data=df)

sb.set(rc={'figure.figsize':(15,5)})
for bars in marital_status.containers:
    marital_status.bar_label(bars)
```

```python
total_sales = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sb.set(rc={'figure.figsize':(6,5)})
sb.barplot(x='Marital_Status', y='Amount', hue='Gender', data = total_sales)
```

<font size='6'>Occupation</font>

```python
total_sales = sb.countplot(x='Occupation', data = df)
sb.set(rc={'figure.figsize': (20,5)})

for bars in total_sales.containers:
    total_sales.bar_label(bars)
```

```python
total_amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head()

sb.barplot(x='Occupation', y='Amount', data=total_amount)
```

<font size='6'>Product Category</font>

```python
df.columns
```

```python
sb.set(rc={'figure.figsize':(25,5)})
pd = sb.countplot(x='Product_Category', data = df)

for bars in pd.containers:
    pd.bar_label(bars)
```

```python
sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(x='Product_Category', y='Amount', data=sales)
```
